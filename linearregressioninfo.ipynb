# Linear Regression — Complete Notes

## 1. Data Representation

Regression problem:

- Dataset with \( n \) samples and \( d \) features

$$
X \in \mathbb{R}^{n \times d}
$$

Each sample:

$$
x_i = [x_{i1}, x_{i2}, \dots, x_{id}]^T
$$

Target:

$$
y_i \in \mathbb{R}
$$

Training dataset:

$$
\{(x_i, y_i)\}_{i=1}^{n}
$$

---

## 2. Goal of ML Model

Find a function:

$$
f: X \rightarrow y
$$

such that:

$$
f(x_i) \approx y_i
$$

---

## 3. Linear Regression Model

### Scalar form:

$$
\hat{y}_i = w_1 x_{i1} + w_2 x_{i2} + \dots + w_d x_{id} + w_0
$$

### Summation form:

$$
\hat{y}_i = \sum_{j=1}^{d} w_j x_{ij} + w_0
$$

### Vector form:

$$
\hat{y}_i = \mathbf{w}^T x_i + w_0
$$

---

## 4. Interpretation

- Analogous to:

$$
y = mx + c
$$

- Finds a **D-dimensional hyperplane** that fits the data

---

## 5. Loss Function (Mean Squared Error)

$$
\min_{w, w_0} \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Expanded:

$$
L(w, w_0) = \frac{1}{n} \sum_{i=1}^{n} \left(y_i - (w^T x_i + w_0)\right)^2
$$

---

## 6. Gradient Computation

### Gradient w.r.t weights:

$$
\frac{\partial L}{\partial w} =
\frac{2}{n} \sum_{i=1}^{n} \left(y_i - (w^T x_i + w_0)\right)(-x_i)
$$

---

### Gradient w.r.t bias:

$$
\frac{\partial L}{\partial w_0} =
\frac{2}{n} \sum_{i=1}^{n} \left(y_i - (w^T x_i + w_0)\right)(-1)
$$

---

## 7. Gradient Descent Update

$$
w = w - \alpha \frac{\partial L}{\partial w}
$$

$$
w_0 = w_0 - \alpha \frac{\partial L}{\partial w_0}
$$

---

## 8. Learning Rate (\(\alpha\))

- Controls step size in optimization

### Effects:
- Small \( \alpha \) → slow convergence
- Large \( \alpha \) → overshooting / divergence

---

## 9. Optimization Goal

Model is optimized when:

- Loss stops decreasing
- Reaches (local/global) minimum

---

## 10. Baseline Model

Mean model:

$$
\hat{y} = \bar{y}
$$

---

## 11. R-Squared Metric

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

Where:

$$
SS_{res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
SS_{tot} = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

---

### Properties:

- Best value:

$$
R^2 = 1 \quad \text{(perfect prediction)}
$$

- Worst case:

$$
R^2 \to -\infty
$$

---

## 12. Adjusted R-Squared

$$
\text{Adj } R^2 = 1 - \left[ \frac{(1 - R^2)(n - 1)}{(n - d - 1)} \right]
$$

Where:
- \( n \): number of samples
- \( d \): number of features

---

## 13. Feature Importance

- Higher weight \( |w_j| \) → more important feature

### Sign of weights:
- \( w_j > 0 \): feature ↑ → prediction ↑  
- \( w_j < 0 \): feature ↑ → prediction ↓  

---

## 14. Feature Scaling (Standardization)

Why?

- Bring all features to same scale
- Avoid misleading importance

Ensures:

$$
\text{range}(f_1) \approx \text{range}(f_2)
$$

---

## ✅ Key Takeaways

- Linear Regression fits a line/hyperplane
- Uses MSE loss
- Optimized via Gradient Descent
- Evaluated using \( R^2 \) / Adjusted \( R^2 \)
- Feature scaling is critical
